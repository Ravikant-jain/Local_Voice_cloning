```python
import torch
import soundfile as sf
import time
import os
from tqdm import tqdm
from qwen_tts import Qwen3TTSModel 

# 1. SETUP PATHS (Absolute paths for Windows stability)
local_model_path = r"B:\AI_noob\Voice_cloning\Qwen3-TTS-12Hz-0.6B-Base"
local_ref_audio = r"B:\AI_noob\Voice_cloning\inputs\Tishya.wav"
output_path = r"B:\AI_noob\Voice_cloning\story_output_voice_clone.wav"

# 2. LOAD MODEL WITH OPTIMIZATIONS
print("--- Loading Qwen3-TTS Model ---")
model = Qwen3TTSModel.from_pretrained(
    local_model_path,
    device_map="cuda:0",
    dtype=torch.bfloat16,
    attn_implementation="sdpa", # Standard for Windows users without flash-attn
    local_files_only=True
)

# Optional: Further optimize for newer PyTorch versions
# model = torch.compile(model) 

# 3. DEFINE INFERENCE TASK
text_to_generate = """ In the quiet town of Chandrapur, where temple bells and evening azaan shared the same sky, lived Aarav and Meera. Aarav measured time by railway whistles, Meera by the rhythm of her anklets as she walked to the old banyan tree every evening. They were different, yet somehow perfectly in step. Every morning, Aarav would leave a small marigold on Meeras windowsill before catching the 7:10 train. Every evening, Meera would light a diya and place it therenever missing a day. No promises were spoken, no letters exchanged. Just faith in small rituals. Years later, when life rushed faster than the trains Aarav once chased, they stood together in the same courtyard. The marigolds were replaced by wedding garlands, the diya by a thousand lamps. Someone asked Aarav the secret of their love.He smiled and said, We never tried to be extraordinary. We just showed up for each otherevery single day. And in that simple devotion, their forever was born."""
ref_text = (
    "So Hi, Im Tishya, a final-year Communication Design student at ISDI. "
    "Ive always been fascinated by doctorsthey never prescribe medicine just because "
    "the packaging looks nice; they dig deep to find the root cause of the pain. "
    "One day, I realized that Design should be exactly the same. "
    "That led me to my core philosophy: 'Diagnose Before Design.' I don't just open a "
    "canvas and start making things"
)

# 4. START TRACKED INFERENCE
print(f"--- Starting Inference on {torch.cuda.get_device_name(0)} ---")

# Using inference_mode for faster, memory-efficient generation
with torch.inference_mode():
    start_time = time.time()
    
    # Simple manual 'infinite loop' guard: wrap in a try/except
    try:
        # We use tqdm here as a visual heartbeat
        with tqdm(total=1, desc="Processing Audio", bar_format="{l_bar}{bar}| {n_fmt}/{total_fmt}") as pbar:
            wavs, sr = model.generate_voice_clone(
                text=text_to_generate,
                language="English",
                ref_audio=local_ref_audio,
                ref_text=ref_text,
                # If your version supports it, add max_new_tokens to prevent stalls
                # max_new_tokens=2048 
            )
            pbar.update(1)
            
        end_time = time.time()
        inference_duration = end_time - start_time
        print(f"\n SUCCESS: Generation took {inference_duration:.2f} seconds.")
        
    except Exception as e:
        print(f"\n ERROR during inference: {e}")
        wavs = None

# 5. SAVE AND LOG RESULT
if wavs is not None:
    sf.write(output_path, wavs[0], sr)
    print(f" Audio saved to: {output_path}")

# Check GPU Memory after run
mem_used = torch.cuda.memory_allocated(0) / 1024**2
print(f" Final VRAM Usage: {mem_used:.2f} MB")
```
